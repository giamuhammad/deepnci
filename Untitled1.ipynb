```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load data
df = pd.read_csv('molecular_properties_results.csv')

# 2. Basic info
print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())
print("\nData types:")
print(df.dtypes)
print("\nMissing values:")
print(df.isnull().sum())

# 3. Descriptive statistics
print("\nDescriptive statistics:")
print(df.describe())

# 4. Extract molecule type from 'molecule_name'
df['molecule_type'] = df['molecule_name'].str.split('_').str[1].str.rstrip('0123456789')

# 5. Distribution of total energy by molecule type
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='molecule_type', y='total_energy')
plt.title('Distribution of Total Energy by Molecule Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 6. Correlation heatmap (numerical columns only)
numerical_cols = df.select_dtypes(include=[np.number]).columns
plt.figure(figsize=(12, 10))
corr = df[numerical_cols].corr()
sns.heatmap(corr, annot=False, cmap='coolwarm', center=0)
plt.title('Feature Correlation Heatmap')
plt.tight_layout()
plt.show()

# 7. Scatter plot: HOMO vs LUMO (if available)
if 'homo_energy' in df.columns and 'lumo_energy' in df.columns:
    # Only non-zero entries have meaningful HOMO/LUMO
    df_hl = df[(df['homo_energy'] != 0) | (df['lumo_energy'] != 0)]
    if not df_hl.empty:
        plt.figure(figsize=(8, 6))
        plt.scatter(df_hl['homo_energy'], df_hl['lumo_energy'], c=df_hl['homo_lumo_gap'], cmap='viridis')
        plt.colorbar(label='HOMO-LUMO Gap')
        plt.xlabel('HOMO Energy (a.u.)')
        plt.ylabel('LUMO Energy (a.u.)')
        plt.title('HOMO vs LUMO Energy (Colored by Gap)')
        plt.tight_layout()
        plt.show()

# 8. Dipole moment magnitude
df['dipole_magnitude'] = np.sqrt(df['dipole_x']**2 + df['dipole_y']**2 + df['dipole_z']**2)

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='molecule_type', y='dipole_magnitude')
plt.title('Dipole Moment Magnitude by Molecule Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 9. Energy components breakdown (example for one molecule type)
sample_mol = df['molecule_type'].unique()[0]
df_sample = df[df['molecule_type'] == sample_mol]
energy_cols = ['potential_energy', 'kinetic_energy', 'correlation_energy', 'exchange_energy']
df_energy = df_sample[energy_cols].reset_index(drop=True)

plt.figure(figsize=(10, 6))
df_energy.plot(kind='line', marker='o')
plt.title(f'Energy Components for {sample_mol} Configurations')
plt.xlabel('Configuration Index')
plt.ylabel('Energy (a.u.)')
plt.legend(title='Energy Type')
plt.tight_layout()
plt.show()
```
